```python
dbscan_mnist_results, dbscan_mnist_best = find_best_dbscan_eps(
    dataset_name="MNIST",
    X=X_mnist,
    y=y_mnist,
    min_samples=10,
    max_points=1000
)
```

```python
summary_rows = []

items = [
    ("Breast Cancer", kmeans_bc_results, dbscan_bc_best),
    ("Adult Income", kmeans_adult_results, dbscan_adult_best),
    ("Covertype", kmeans_cover_results, dbscan_cover_best),
    ("MNIST", kmeans_mnist_results, dbscan_mnist_best),
]

for dataset_name, kmeans_result, dbscan_best in items:
    best_kmeans = kmeans_result.loc[kmeans_result["ARI"].idxmax()]

    kmeans_ari = best_kmeans["ARI"]
    dbscan_ari = dbscan_best["ARI"]

    if kmeans_ari > dbscan_ari:
        better_method = "K-Means"
    elif dbscan_ari > kmeans_ari:
        better_method = "DBSCAN"
    else:
        better_method = "Tie"

    summary_rows.append({
        "Dataset": dataset_name,
        "Best KMeans k": int(best_kmeans["k"]),
        "Best KMeans ARI": round(kmeans_ari, 4),
        "Best DBSCAN eps": round(dbscan_best["eps"], 4),
        "DBSCAN clusters": int(dbscan_best["clusters"]),
        "DBSCAN ARI": round(dbscan_ari, 4),
        "DBSCAN noise fraction": round(dbscan_best["noise_fraction"], 4),
        "Better method": better_method
    })

clustering_summary = pd.DataFrame(summary_rows)
clustering_summary
```
